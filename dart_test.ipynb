```python
from bs4 import BeautifulSoup

def parse_contract_info(html: str) -> dict:
    soup = BeautifulSoup(html, 'html.parser')
    table = soup.find('table')
    # 뽑고 싶은 라벨 ↔ 딕셔너리 키 매핑
    labels = {
        '체결계약명': 'contract_name',
        '계약금액(원)': 'contract_amount',
        '최근매출액(원)': 'recent_sales_amount',
        '매출액대비(%)': 'sales_ratio_percent',
        '계약(수주)일자': 'contract_date'
    }
    info = {}
    
    for tr in table.find_all('tr'):
        tds = tr.find_all('td')
        if len(tds) < 2:
            continue

        # <td> 3개짜리(헤더가 2칸에 나뉘어 있는 경우)
        if len(tds) >= 3:
            header_text = ''.join(td.get_text(strip=True) for td in tds[:2])
            value_td    = tds[2]
        # <td> 2개짜리(헤더가 colspan=2인 경우)
        else:
            header_text = tds[0].get_text(strip=True)
            value_td    = tds[1]

        # 각 라벨별로 매칭
        for label, key in labels.items():
            if label in header_text:
                # 값 추출
                span = value_td.find('span', class_='xforms_input') or value_td.find('span')
                raw  = span.get_text(strip=True) if span else value_td.get_text(strip=True)

                # 타입 변환
                if key in ('contract_amount', 'recent_sales_amount'):
                    val = int(raw.replace(',', ''))
                elif key == 'sales_ratio_percent':
                    val = float(raw)
                else:
                    val = raw

                info[key] = val
                break

    return info

# ──────────────────────────────────────────────────────────────────────────────

# 사용 예시 (위에서 가져온 full_html이 있을 때)
parsed = parse_contract_info(full_text)
print(parsed)
```

```python
import requests
from io import BytesIO
import pandas as pd
import re
from bs4 import BeautifulSoup

def get_biz_day():
    url = 'https://finance.naver.com/sise/sise_deposit.nhn'
    res = requests.get(url)
    txt = BeautifulSoup(res.text, 'lxml').select_one('span.tah').text
    return ''.join(re.findall(r'\d+', txt))

def fetch_market(mkt_id, biz_day, sess):
    otp = sess.post(
        "http://data.krx.co.kr/comm/fileDn/GenerateOTP/generate.cmd",
        data={
            "locale":"ko_KR","mktId":mkt_id,"trdDd":biz_day,
            "money":"1","csvxls_isNo":"false","name":"fileDown",
            "url":"dbms/MDC/STAT/standard/MDCSTAT03901"
        }
    ).text.strip()
    csv_resp = sess.post(
        "http://data.krx.co.kr/comm/fileDn/download_csv/download.cmd",
        data={"code": otp},
    )
    df = pd.read_csv(BytesIO(csv_resp.content), encoding="EUC-KR")
    df['기준일'] = biz_day
    df['시장']   = 'KOSPI' if mkt_id=='STK' else 'KOSDAQ'
    return df[['시장','종목코드','종목명','업종명','시가총액','기준일']]

sess = requests.Session()
sess.headers.update({
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Referer":"http://data.krx.co.kr/contents/MDC/MDI/mdiLoader/index.cmd?menuId=MDC0201020506",
    "X-Requested-With":"XMLHttpRequest",
})
sess.get(sess.headers['Referer'])

biz_day = get_biz_day()
markets = ['STK','KSQ']
df_list = [fetch_market(m, biz_day, sess) for m in markets]
result = pd.concat(df_list, ignore_index=True)
```

### 네이버페이 증권 
- 시가총액, WICS 업종 분류

```python
import requests
from bs4 import BeautifulSoup

def fetch_wics_and_marketcap(code: str) -> dict:
    """
    종목코드(code) 하나로 WICS 업종명과 시가총액을
    navercomp API 응답에서 동시에 추출하여 반환합니다.
    """
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Referer":    f"https://finance.naver.com/item/main.nhn?code={code}"
    }
    url = f"https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={code}&cn="
    res = requests.get(url, headers=headers)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, "lxml")

    # 1) WICS 추출
    wics = None
    for dt in soup.select("td.td0101 dl dt.line-left"):
        text = dt.get_text(strip=True)
        if text.startswith("WICS"):
            # "WICS : 기계" → "기계"
            wics = text.split(":", 1)[1].strip()
            break

    # 2) 시가총액 추출
    mktcap = None
    table = soup.find("table", id="cTB11")
    if table:
        for tr in table.find_all("tr"):
            th = tr.find("th", class_="txt")
            if th and "시가총액" in th.get_text():
                td = tr.find("td", class_="num")
                mktcap = td.get_text(strip=True)
                break

    return {"WICS": wics, "시가총액": mktcap}

if __name__ == "__main__":
    code = "064350"  # 예: 현대로템
    info = fetch_wics_and_marketcap(code)
    print(info)
```

```python
import requests
import zipfile, io
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup
from dotenv import load_dotenv
import os
import re

# API_KEY       = '17db1c341b5874ad5ad8a43382af6b17d3009380'
API_KEY = os.getenv("DART_API_KEY")
TODAY         = datetime.now().strftime('%Y%m%d')
TARGET_STOCKS = ['071970', '064350', '010960']

# --- 0) Session 생성 (커넥션 풀링) ---
session = requests.Session()
session.headers.update({"User-Agent": "Mozilla/5.0"})

def fetch_today_single_sales() -> pd.DataFrame:
    """오늘자 단일판매·공급계약 공시를 DataFrame으로 반환합니다."""
    url = 'https://opendart.fss.or.kr/api/list.json'
    params = {
        'crtfc_key':        API_KEY,
        'bgn_de':           TODAY,
        'end_de':           TODAY,
        'pblntf_ty':        'I',
        'page_count':       300
    }
    resp = requests.get(url, params=params, timeout=10)
    resp.raise_for_status()
    data = resp.json().get('list', [])
    filtered = [d for d in data if '단일판매' in d.get('report_nm','')]
    return pd.DataFrame(filtered)

def parse_contract_html(rcept_no: str) -> dict:
    """rcept_no로 상세공시 HTML을 받아 주요 값을 파싱해 반환합니다."""
    r = requests.get(
        'https://opendart.fss.or.kr/api/document.xml',
        params={'crtfc_key':API_KEY, 'rcept_no':rcept_no},
        timeout=10
    )
    r.raise_for_status()
    z = zipfile.ZipFile(io.BytesIO(r.content))
    html = z.read(next(f for f in z.namelist() if f.lower().endswith(('.xml','.html'))))\
               .decode('utf-8', errors='ignore')
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('table')

    def get_val(keyword):
        for tr in table.find_all('tr'):
            tds = tr.find_all('td')
            if len(tds) < 2: continue
            label = re.sub(r'\s+', '', tds[0].get_text() + tds[1].get_text())
            if keyword in label:
                return tds[-1].get_text(strip=True).replace(',', '')
        return None

    return {
        'contract_name':       get_val('체결계약명'),
        'contract_amount':     int(get_val('계약금액') or 0),
        'recent_sales_amount': int(get_val('최근매출액') or 0),
        'sales_ratio_percent': float(get_val('매출액대비') or 0),
        'contract_date':       get_val('계약(수주)일자')
    }

def fetch_market_info(session: requests.Session, stock_code: str) -> dict:
    headers = {"Referer": f"https://finance.naver.com/item/main.nhn?code={stock_code}"}
    url = f"https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={stock_code}&cn="
    res = session.get(url, headers=headers, timeout=10)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')

    # WICS
    wics = None
    for dt in soup.select("td.td0101 dl dt.line-left"):
        txt = dt.get_text(strip=True)
        if txt.startswith("WICS"):
            wics = txt.split(":",1)[1].strip()
            break

    # 시가총액
    mktcap = None
    tbl = soup.find("table", id="cTB11")
    if tbl:
        for tr in tbl.find_all("tr"):
            th = tr.find("th", class_="txt")
            if th and "시가총액" in th.get_text():
                mktcap = tr.find("td", class_="num").get_text(strip=True)
                break

    return {'WICS': wics, 'market_cap': mktcap}
```

```python
from dotenv import load_dotenv
import os
import re
import zipfile
import io
import requests
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup

load_dotenv()
API_KEY = os.getenv("DART_API_KEY")
TODAY   = datetime.now().strftime('%Y%m%d')
# TODAY = '20250526'

session = requests.Session()
session.headers.update({"User-Agent": "Mozilla/5.0"})

def fetch_sales(session) -> pd.DataFrame:
    url = 'https://opendart.fss.or.kr/api/list.json'
    params = {
        'crtfc_key': API_KEY, 'bgn_de': TODAY, 'end_de': TODAY,
        'pblntf_detail_ty':'I001', 'page_count':100, 'last_reprt_at':'Y', 
    }
    resp = session.get(url, params=params, timeout=10)
    resp.raise_for_status()
    data = resp.json().get('list', [])
    filtered = [
        d for d in data
        if '단일판매' in d.get('report_nm','')
        and '정정' not in d.get('report_nm','')
        and '해지' not in d.get('report_nm','')
    ]
    return pd.DataFrame(filtered)

def parse_contract(session, rcept_no: str) -> dict:
    """rcept_no로 상세공시 ZIP을 받아, 정정·일반 보고서 테이블 모두 대응해 파싱."""
    # 1) document.xml API 호출 & HTML 디코딩 (기존과 동일)
    resp = session.get(
        'https://opendart.fss.or.kr/api/document.xml',
        params={'crtfc_key': API_KEY, 'rcept_no': rcept_no},
        timeout=10
    )
    resp.raise_for_status()
    z = zipfile.ZipFile(io.BytesIO(resp.content))
    fname = next(f for f in z.namelist() if f.lower().endswith(('.xml','.html')))
    html = z.read(fname).decode('utf-8', errors='ignore')
    soup = BeautifulSoup(html, 'lxml')

    # 2) 올바른 테이블 선택
    contract_table = None
    for tbl in soup.find_all('table'):
        text = tbl.get_text()
        if '체결계약명' in text or '계약금액' in text or '판매ㆍ공급계약 내용' in text:
            contract_table = tbl
            break
    if contract_table is None:
        raise RuntimeError("계약 테이블을 찾을 수 없습니다.")

    # 3) 값 추출 헬퍼
    def get_val(keys):
        for tr in contract_table.find_all('tr'):
            tds = tr.find_all('td')
            if len(tds) < 2:
                continue
            label = re.sub(r'\s+', '', tds[0].get_text() + tds[1].get_text())
            if any(k.replace('ㆍ','') in label or k in label for k in keys):
                return tds[-1].get_text(strip=True).replace(',', '')
        return None

    # 4) 필드별 파싱 (contract_name에 fallback 추가)
    name = get_val(['체결계약명'])
    if not name:
        # 정정보고 등에서 '판매ㆍ공급계약 내용'에 계약명이 있을 때
        name = get_val(['판매ㆍ공급계약내용', '판매ㆍ공급계약 내용'])

    raw_start = get_val(['시작일'])
    # raw_start = get_val(['계약(수주)일자'])
    if raw_start in (None, '', '-'):
        raw_start = get_val(['계약(수주)일자'])
    
    return {
        '내용 ':       name or '',
        '계약 금액(억)':     int(get_val(['계약금액','계약금액총액']) or 0)/100_000_000,
        # 'recent_sales_amount': int(get_val(['최근매출액','최근매출액총액']) or 0),
        '매출액 대비(%) (A)': float(get_val(['매출액대비','매출액대비(%)']) or 0),
        '시작일 (s)':       raw_start,
        '종료일 (e)':       get_val(['종료일']),
        '계약상대':        get_val(['계약상대']),
    }

def fetch_market_info(session, stock_code: str) -> dict:
    url  = f"https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={stock_code}&cn="
    res  = session.get(url, timeout=10)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')

    # WICS
    wics = None
    for dt in soup.select("td.td0101 dl dt.line-left"):
        txt = dt.get_text(strip=True)
        if txt.startswith("WICS"):
            wics = txt.split(":",1)[1].strip()
            break

    # 시가총액 → 숫자만
    mktcap = None
    tbl    = soup.find("table", id="cTB11")
    if tbl:
        for tr in tbl.find_all("tr"):
            th = tr.find("th", class_="txt")
            if th and "시가총액" in th.get_text():
                raw = tr.find("td", class_="num").get_text(strip=True)
                # e.g. "244억원" or "1,318,44억원"
                num_str = raw.replace("억원","").replace(",","")
                try:
                    mktcap = int(num_str)
                except ValueError:
                    mktcap = None
                break

    return {'업종 분류': wics, '시가총액(억)': mktcap}
```

```python
from dotenv import load_dotenv
import os
import re
import zipfile
import io
import requests
import pandas as pd
from datetime import datetime
from bs4 import BeautifulSoup

load_dotenv()
API_KEY = os.getenv("DART_API_KEY")
TODAY = datetime.now().strftime('%Y%m%d')
# TODAY = '20250507'

session = requests.Session()
session.headers.update({"User-Agent": "Mozilla/5.0"})

def fetch_sales(session) -> pd.DataFrame:
    url = 'https://opendart.fss.or.kr/api/list.json'
    base_params = {
        'crtfc_key': API_KEY,
        'bgn_de': TODAY,
        'end_de': TODAY,
        'pblntf_detail_ty': 'I001',
        'page_count': 100,
        'last_reprt_at': 'Y',
    }

    all_reports = []
    params = base_params.copy()
    params['page_no'] = 1
    resp = session.get(url, params=params, timeout=10)
    resp.raise_for_status()
    data = resp.json()
    total_page = int(data.get('total_page', 1))
    all_reports.extend(data.get('list', []))

    for page in range(2, total_page + 1):
        params = base_params.copy()
        params['page_no'] = page
        resp = session.get(url, params=params, timeout=10)
        resp.raise_for_status()
        reports = resp.json().get('list', [])
        if not reports:
            break
        all_reports.extend(reports)

    df = pd.DataFrame(all_reports)
    filtered = df[
        df['report_nm'].str.contains('단일판매') &
        ~df['report_nm'].str.contains('정정|해지')
    ].reset_index(drop=True)

    return filtered

def parse_contract(session, rcept_no: str) -> dict:
    resp = session.get(
        'https://opendart.fss.or.kr/api/document.xml',
        params={'crtfc_key': API_KEY, 'rcept_no': rcept_no},
        timeout=10
    )
    resp.raise_for_status()

    # 1) ZIP 여부 판별 & 읽기
    try:
        z = zipfile.ZipFile(io.BytesIO(resp.content))
    except zipfile.BadZipFile:
        # ZIP 파일이 아니면 그냥 무시
        return {}

    fname = next(f for f in z.namelist() if f.lower().endswith(('.xml', '.html')))
    html = z.read(fname).decode('utf-8', errors='ignore')
    soup = BeautifulSoup(html, 'lxml')

    # 2) 계약 테이블 찾기
    contract_table = None
    for tbl in soup.find_all('table'):
        text = tbl.get_text()
        if '체결계약명' in text or '계약금액' in text or '판매ㆍ공급계약 내용' in text:
            contract_table = tbl
            break
    if contract_table is None:
        raise RuntimeError("계약 테이블을 찾을 수 없습니다.")

    # 3) 값 추출 헬퍼
    def get_val(keys):
        for tr in contract_table.find_all('tr'):
            tds = tr.find_all('td')
            if len(tds) < 2:
                continue
            label = re.sub(r'\s+', '', tds[0].get_text() + tds[1].get_text())
            if any(k.replace('ㆍ', '') in label or k in label for k in keys):
                return tds[-1].get_text(strip=True).replace(',', '')
        return None

    def get_int(keys):
        raw = get_val(keys)
        if raw in (None, '', '-'):
            return None
        return int(raw)

    def get_float(keys):
        raw = get_val(keys)
        if raw in (None, '', '-'):
            return None
        return float(raw)

    # 4) 필드 매핑
    name = get_val(['체결계약명']) or get_val(['판매ㆍ공급계약내용', '판매ㆍ공급계약 내용']) or get_val(['- 체결계약명'])
    raw_start = get_val(['시작일'])
    if raw_start in (None, '', '-'):
        raw_start = get_val(['계약(수주)일자'])

    return {
        '내용':            name or '',
        '계약 금액(억)':    int(get_int(['계약금액', '계약금액총액', '계약금액(원원)']) or 0) / 100_000_000,
        '매출액 대비(%) (A)': float(get_float(['매출액대비', '매출액대비(%)']) or 0),
        '시작일 (s)':       raw_start,
        '종료일 (e)':       get_val(['종료일']),
        '계약상대':          get_val(['계약상대']),
    }


def fetch_market_info(session, stock_code: str) -> dict:
    url = f"https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd={stock_code}&cn="
    res = session.get(url, timeout=10)
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')

    wics = None
    for dt in soup.select("td.td0101 dl dt.line-left"):
        txt = dt.get_text(strip=True)
        if txt.startswith("WICS"):
            wics = txt.split(":", 1)[1].strip()
            break

    mktcap = None
    tbl = soup.find("table", id="cTB11")
    if tbl:
        for tr in tbl.find_all("tr"):
            th = tr.find("th", class_="txt")
            if th and "시가총액" in th.get_text():
                raw = tr.find("td", class_="num").get_text(strip=True)
                num_str = raw.replace("억원", "").replace(",", "")
                try:
                    mktcap = int(num_str)
                except ValueError:
                    mktcap = None
                break

    return {'업종 분류': wics, '시가총액(억)': mktcap}
```

```python
import os
import pandas as pd
import numpy as np
from openpyxl import load_workbook, Workbook
from openpyxl.styles import Alignment
from zipfile import BadZipFile

if __name__ == '__main__':
    df = fetch_sales(session)
    if df.empty:
        print("오늘 단일판매 공시가 없습니다.")
    else:
        mapping = {'Y': 'KS', 'K': 'KQ'}
        codes = df['stock_code'].astype(str).unique()
        market_infos = {code: fetch_market_info(session, code) for code in codes}

        records = []
        for _, row in df.iterrows():
            detail = parse_contract(session, row['rcept_no'])
            market = market_infos[str(row['stock_code'])]
            if market.get('업종 분류') == '건설':
                continue
            records.append({
                '공시회사': row['corp_name'],
                '날짜 (D)': row['rcept_dt'],
                '거래소': mapping.get(row['corp_cls'], ''),
                '내용': detail.get('내용', ''),
                '계약 금액(억)': detail.get('계약 금액(억)', 0.0),
                '매출액 대비(%) (A)': detail.get('매출액 대비(%) (A)', 0.0),
                '계약상대': detail.get('계약상대', ''),
                '시작일 (s)': detail.get('시작일 (s)'),
                '종료일 (e)': detail.get('종료일 (e)'),
                '업종 분류': market.get('업종 분류', ''),
                '시가총액(억)': market.get('시가총액(억)', 0)
            })

        result_df = pd.DataFrame(records)
        result_df['날짜 (D)'] = pd.to_datetime(result_df['날짜 (D)'], format='%Y%m%d', errors='coerce')
        result_df['시작일 (s)'] = pd.to_datetime(result_df['시작일 (s)'], format='%Y-%m-%d', errors='coerce')
        result_df['종료일 (e)'] = pd.to_datetime(result_df['종료일 (e)'], format='%Y-%m-%d', errors='coerce')

        if result_df['계약 금액(억)'].dtype == object:
            result_df['계약 금액(억)'] = result_df['계약 금액(억)'].astype(str).str.replace(',', '').astype(float)
        if result_df['매출액 대비(%) (A)'].dtype == object:
            result_df['매출액 대비(%) (A)'] = result_df['매출액 대비(%) (A)'].astype(str).str.replace(',', '').astype(float)
        if result_df['시가총액(억)'].dtype == object:
            result_df['시가총액(억)'] = result_df['시가총액(억)'].astype(str).str.replace(',', '').astype(int)

        excel_path = '국내 주요 공시 정리.xlsx'
        if os.path.exists(excel_path):
            try:
                wb = load_workbook(excel_path)
            except BadZipFile:
                wb = Workbook()
        else:
            wb = Workbook()

        if 'main' in wb.sheetnames:
            ws = wb['main']
            existing_df = pd.read_excel(excel_path, sheet_name='main')
        else:
            ws = wb.create_sheet('main')
            ws.append(list(result_df.columns))
            existing_df = pd.DataFrame(columns=result_df.columns)

        header_row = [cell.value for cell in ws[1]]
        numeric_cols = result_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
        col_idx_map = {col: header_row.index(col) + 1 for col in numeric_cols if col in header_row}

        merge_cols = ['날짜 (D)', '공시회사', '계약 금액(억)']
        if not existing_df.empty:
            merged = result_df.merge(existing_df, on=merge_cols, how='left', indicator=True)
            new_rows = merged[merged['_merge'] == 'left_only'].drop(columns=['_merge'])
        else:
            new_rows = result_df.copy()

        for _, row in new_rows.iterrows():
            row_values = [row.get(col, '') for col in header_row]
            ws.append(row_values)
            new_row_idx = ws.max_row

            for idx, col in enumerate(header_row, start=1):
                cell = ws.cell(row=new_row_idx, column=idx)
                value = row.get(col, '')
                if col in ['날짜 (D)', '시작일 (s)', '종료일 (e)'] and pd.notna(value):
                    cell.number_format = 'yyyy-mm-dd'
                elif col in col_idx_map:
                    if np.issubdtype(result_df[col].dtype, np.integer):
                        cell.number_format = '#,##0'
                    else:
                        cell.number_format = '#,##0.00'
                if col not in ['내용', '계약상대']:
                    cell.alignment = Alignment(horizontal='center')

        wb.save(excel_path)
        print("✅ 공시 업데이트 완료")
```
